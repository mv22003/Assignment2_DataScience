```python
 # Collect starting and ending point of stimulus
startPoints = df[df['Event']=='ImageStimulusStart'].index.values.astype(int)
endPoints = df[df['Event']=='ImageStimulusEnd'].index.values.astype(int)
startPoints
```

```python
new_df = pd.DataFrame()

# Store only image stimulus
for i in range(len(startPoints)):
    start = startPoints[i]
    end = endPoints[i]

    trial = df.iloc[start:end+1]
    new_df = pd.concat([new_df,trial])
```
